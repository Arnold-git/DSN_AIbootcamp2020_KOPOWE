```python
from sklearn.model_selection import KFold, StratifiedKFold, StratifiedShuffleSplit
from sklearn.metrics import roc_auc_score
from sklearn.metrics import log_loss
from catboost import CatBoostClassifier

# pred_score = []
# test_pred = []
splits = 20
# errlg = []
y_test_pred = 0
fold = StratifiedKFold(n_splits = splits, shuffle = True, random_state = 2020)

for train_index, test_index in fold.split(X, y):
  X_train, X_test = X.loc[train_index], X.loc[test_index]
  y_train, y_test = y.loc[train_index], y.loc[test_index]

  cat = CatBoostClassifier(task_type="GPU", eval_metric="AUC")
  cat.fit(X_train, y_train, # eval_set=[(X_train, y_train), (X_test, y_test)],
          verbose = True,
          )
  
  pred = cat.predict_proba(X_test)[:, 1]
  print( "  auc = ", roc_auc_score(y_test, pred) )
  print('Logloss', log_loss(y_test, pred), "*"*30)
  # errlg.append(log_loss(y_test, pred_score))
  y_test_pred += cat.predict_proba(test[X.columns])[:, 1]
  # test_pred.append(p)
y_test_pred /= splits
```
